The following code shows an example of how to animate the gradient descent algorithm. By changing the "learning rate" parameter, it can be seen how the algorithm finds its way to the minimum! This example demonstrates on the function $f(x) = x^2 + 0.5x + 3.0$.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Define the function to minimize (example: a simple quadratic function)
def f(x):
    return x**2 + 0.5*x +3
# Define the gradient of the function (derivative)
def gradient(x):
    return 2 * x + 0.5

# Define the gradient descent algorithm
def gradient_descent(learning_rate, num_iterations):
    x = 5  # Initial guess
    x_history = [x]
    
    for i in range(num_iterations):
        x = x - learning_rate * gradient(x)
        x_history.append(x)
    
    return x_history

# Parameters
learning_rate = 0.1
num_iterations = 30

# Run gradient descent
x_history = gradient_descent(learning_rate, num_iterations)

# Create figure and axis
fig, ax = plt.subplots()
ax.set_xlim(-6, 6)
# ax.set_ylim(0, 30)

# Plot the function
x_values = np.linspace(-6, 6, 100)
y_values = f(x_values)
ax.plot(x_values, y_values, color='blue')

# Initialize empty plot for animation
line, = ax.plot([], [], 'ro')

# Update function for animation
def update(frame):
    line.set_data(x_history[frame], f(x_history[frame]))
    return line,

# Create animation
ani = FuncAnimation(fig, update, frames=len(x_history), blit=True)

plt.title('Gradient Descent Animation (Learning Rate: {})'.format(learning_rate))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()

ani.save("gradient_descent3.mp4", fps = 3)
```
